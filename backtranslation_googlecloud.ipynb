```python
import pandas as pd

df = pd.read_csv("klue-xnli-hypo_gen_include.csv", index_col="index")
```

```python
from google.cloud import translate_v2 as translate

translate_client = translate.Client()
```

```python
import six
```

```python
def translate_to_en(text):
    if isinstance(text, six.binary_type):
        text = text.decode("utf-8")

    eng = translate_client.translate(text, target_language="en")

    return eng["translatedText"]
```

```python
def translate_to_kr(text):
    if isinstance(text, six.binary_type):
        text = text.decode("utf-8")

    kr = translate_client.translate(text, target_language="ko")

    return kr["translatedText"]
```
